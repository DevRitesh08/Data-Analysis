# Data Analysis

**it is a process of inspecting , cleansing , transforming and modelling data with a goal of discovering useful information , informing conclusions and supporting decision-making**



---

## Step 1. Asking Questions

1. What features will contribute to my analysis?
2. What features are not important for my analysis?
3. Which of the features have a strong correlation?
4. Do I need data preprocessing?
5. What kind of feature manipulation/engineering is required?

## Step 2. Data Wrangling
Data wrangling, sometimes referred to as data munging, is the process of transforming and mapping data from one "raw" data form into another format with the intent of making it more appropriate and valuable for a variety of downstream purposes such as analytics.

it can be further divided into three parts :
1. Data Gathering
2. Data Assessing
3. Data Cleaning

### 1. Data Gathering :
It includes collecting data from various sources like csv files, api , web scraping and databases .

### 2. Assessing Data :
1. Finding the number of rows/columns (shape)
2. Data types of various columns (info())
3. Checking for missing values (info())
4. Check for duplicate data (is_unique)
5. Memory occupied by the dataset (info)
6. High level mathematical overview of the data (describe)

### 3. Data Cleaning
1. Missing Data (e.g. mean)
2. Remove duplicate data (drop_duplicates)
3. Incorrect data type (astype)

## Step 3. Exploratory Data Analysis
Exploratory Data Analysis (EDA) is the process of systematically exploring a dataset to uncover its underlying structure, patterns, and relationships. It often involves visualizing, summarizing, and augmenting the data to gain valuable insights, detect anomalies, and prepare it for further analysis or modeling.

    1. Exploring the data :
        - finding correlation and covariance .
        - Doing univariate and multivariate analysis .
        - Plotting Graphs ( Data Visualization )
    2. Augmenting Data
        - removing outliers
        - merging dataframes
        - adding new columns

    These Operations are collectively known as feature engineering .


`Feature engineering : is the process of using domain knowledge to select, transform, create, or extract new features (variables or attributes) from raw data to improve the performance of machine learning models.`

## Step 4. Drawing Conclusions
This can be done using Machine Learning , Inferential Statistics and Descriptive Statistics .

#### Some example conclusions based on Descriptive Statistics

1. Is Rohit Sharma a better batsman in 2nd innings (IPL Dataset)?
2. Does being a female increase your chances of survival (Titanic Dataset)?
3. Is Delhi the most costly place for eating out (Zomato Dataset)?

`Descriptive statistics are statistical methods that summarize and organize the main features of a dataset, providing simple quantitative descriptions. They help present large amounts of data in a meaningful way, making it easier to understand patterns, trends, and distributions.`

`Descriptive statistics do not make predictions or infer properties about a larger population; instead, they describe the characteristics of the data at hand , hence it is different from inferential statistics and machine learning .`
